```python
import requests
import pandas as pd
import json

url = ('https://api.thegraph.com/subgraphs/name/itsjerryokolo/cryptopunks')

query = '''
{sales(first: 1000){
    id
    to {
      id
    }
    amount
    from {
      id
    }
    nft {
      id
    }
    timestamp
    type
    contract {
      id
    }}}
  '''

response = requests.post(url, json={'query': query})

print(f"Request returned {response.status_code} : '{response.reason}'")

payload = response.json()

# testing the response to make sure the results are showing up
import pprint
pp = pprint.PrettyPrinter(indent=1)
pp.pprint(payload['data']['sales'][0:5])
```

```python
df = pd.json_normalize(payload['data']['sales'])
```
